# Encoding categorical data

```python
# Label Encoder

LEncoder = LabelEncoder()

df['diagnosis'] = LEncoder.fit_transform(df['diagnosis'])
```

# Pre-Modeling Tasks

## Separating the independant and the dependant variable

```python
X = df[cols]
X = df.drop('diagnosis',axis=1)
y = df['diagnosis'].values
```

## Splitting the dataset

```python
random_state = 42

x_train, x_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=random_state)
```

# Feature Scaling

```python
sc = StandardScaler()

X_train = sc.fit_transform(x_train)
X_test= sc.transform(x_test)
```

# Modeling

```python
# Logistic Regression


logreg= LogisticRegression()

logreg.fit(X_train, y_train)

y_pred_logreg = logreg.predict(X_test)

# Saving model to disk
pickle.dump(logreg, open('Logi_model.pkl','wb'))


# Random Forest Classifier

rf = RandomForestClassifier()

rf.fit(X_train, y_train)

y_pred_rf = rf.predict(X_test)

pickle.dump(rf, open('RF_model.pkl','wb'))


# Decision Tree Classifier

dt = DecisionTreeClassifier()

dt.fit(X_train, y_train)

y_pred_dt = dt.predict(X_test)

pickle.dump(dt, open('DT_model.pkl','wb'))



# KNeighbors Classifier


knn = KNeighborsClassifier(n_neighbors=5)

knn.fit(X_train, y_train)

y_pred_knn = knn.predict(X_test)

pickle.dump(knn, open('knn_model.pkl','wb'))


# Support Vector classifier

svc = SVC(probability=True)

svc.fit(X_train,y_train)

y_pred_svc = svc.predict(X_test)

pickle.dump(svc, open('svc_model.pkl','wb'))
```

```python
X_train.shape, y_train.shape,X_test.shape, y_test.shape
```

# Evaluation and comparison of all the models

```python
models = []

Z = [SVC() , DecisionTreeClassifier() , LogisticRegression() , KNeighborsClassifier(), RandomForestClassifier()]


X = ["SVC" , "DecisionTreeClassifier" , "LogisticRegression" , "KNeighborsClassifier" , "RandomForestClassifier"]

for i in range(0,len(Z)):
    model = Z[i]
    model.fit( X_train , y_train )
    pred = model.predict(X_test)
    models.append(accuracy_score(pred , y_test))
```

```python
d = { "Accuracy" : models , "Algorithm" : X }
data_frame = pd.DataFrame(d)
data_frame
```

```python
sns.barplot(x=data_frame['Accuracy'],y=data_frame['Algorithm'],palette= "husl").set_title('Accuracy of all Algorithms')
plt.show()
```
